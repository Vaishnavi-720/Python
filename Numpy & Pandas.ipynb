## Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
df.species.unique()
```

```python
df.species.value_counts()
```

```python
df.head()
```

# Relational Plots

```python
#line plot
sns.lineplot(data=df, x='sepal_length', y='sepal_width')
plt.show()
```

```python
#line plot
sns.lineplot(x = 'sepal_length', y = 'sepal_width', data = df, hue = 'species', ci= None)
plt.show()
```

## Scattter Plot

```python
#scatter pllot
sns.scatterplot(x = 'petal_width', y = 'petal_length', data = df)
plt.title('Scatter Plot')
plt.show()
```

```python
sns.scatterplot(x = 'petal_width', y = 'petal_length', data = df, hue = 'species')
plt.title('Scatter Plot')
plt.show()
```

## Rel Plot

```python
#relplot by default is scatter plot
#relplot can be converted into plot by giving kind = 'line'
sns.relplot(data = df, x = 'petal_length', y = 'sepal_width')
plt.show()
```

```python
sns.relplot(data=df, x = 'petal_length', y = 'sepal_width', kind = 'line', ci = None)
plt.show()
```

## Categorical Plots


```python
#Count plot value_count()

sns.countplot(x = 'species', data = df)
plt.show()
```

```python
#Bar Plot
sns.barplot(x = 'species', y = 'sepal_length', data = df)
plt.show()
```

```python
df
```

```python
#Box Plot
sns.boxplot(x = 'sepal_width',data = df)
plt.show()
```

```python
sns.boxplot(x = 'species', y = "sepal_width",data = df)
plt.show()
```

```python
#violin plot
sns.violinplot(x='sepal_width', data = df)
plt.show()
```

```python
sns.violinplot(x='species', y= 'sepal_width', data = df)
plt.show()
```

```python
#strip plot
sns.stripplot(x='species', y ='sepal_width', data = df, hue = 'species')
plt.show()
```

```python
sns.swarmplot(x='species', y ='sepal_width', data = df, hue = 'species')
plt.show()
```

```python
#categorical plot
sns.catplot(x='sepal_length', data = df, kind ='violin')
plt.show()
```

```python
sns.catplot(x = 'species', y='sepal_length', data = df, kind ='violin')
plt.show()
```

```python
sns.catplot(x = 'species', y='sepal_length', data = df, kind ='swarm')
plt.show()
```

```python
#catplot is bydefault strip plot
sns.catplot(x = 'petal_length', y='species', data = df)
plt.show()
```
